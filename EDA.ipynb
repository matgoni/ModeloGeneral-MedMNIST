## Coeficiente de silueta

```python
from scripts.silhouette import SilhouetteAnalyzer
```

```python

analyzer = SilhouetteAnalyzer(['bloodmnist','breastmnist', 'chestmnist', 'dermamnist', 'octmnist', 'organamnist', 'organcmnist', 'organsmnist', 'pathmnist', 'pneumoniamnist','retinamnist','tissuemnist'], samples_per_dataset=1500)
score_classes = analyzer.mean_score(label_level="class")
```

```python
print("Silhouette por clase:", score_classes)
```

```python
score_dataset = analyzer.mean_score(label_level="dataset")
print("Silhouette por dataset:", score_dataset)
```
